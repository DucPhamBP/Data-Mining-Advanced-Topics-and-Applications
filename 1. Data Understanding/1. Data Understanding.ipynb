```python
import numpy as np
from scipy import stats

# list of numerical columns to check for outliers
numerical_cols = ['popularity','followers']

# create a copy 
df_aritist_final = dup_a.copy()

# calculate the Z-scores for all numerical columns
z_scores = np.abs(stats.zscore(df_aritist_final[numerical_cols]))

# set the threshold for identifying outliers
threshold = 3

# create a boolean mask where True indicates no outliers
mask = (z_scores < threshold).all(axis=1)

# apply the mask to filter out rows with any outliers
df_aritist_final_cleaned = df_aritist_final[mask]

df_aritist_final_cleaned
```

```python
numerical_cols_tracks= ["popularity", "followers"]
for col in numerical_cols_tracks:
    z = np.abs(stats.zscore(dup_a[numerical_cols_tracks]))
    threshold = 3
    print(col,np.where(z > 3))
```

```python
import pandas as pd

# calculate popularity and followers
def calculate_popularity_followers(row):
    artists = row['artists'].split(';')
    if len(artists) == 1:
        artist_data = dup_a[dup_a['name'] == artists[0]]
        if not artist_data.empty:
            return artist_data['popularity'].values[0], artist_data['followers'].values[0]
    elif len(artists) > 1:
        popularity_sum = 0
        followers_sum = 0
        for artist in artists:
            artist_data = dup_a[dup_a['name'] == artist]
            if not artist_data.empty:
                popularity_sum += artist_data['popularity'].values[0]
                followers_sum += artist_data['followers'].values[0]
        return popularity_sum / len(artists), followers_sum / len(artists)
    return None, None  # Return None if artist not found or no artists listed

# calculate popularity and followers
dup_t['popularity_art'], dup_t['followers'] = zip(*dup_t.apply(calculate_popularity_followers, axis=1))
print(dup_t)
```
